```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import linregress
```

### 1. Encontrar el valor atípico y explicar por qué se considera un valor extraño.

El valor atípico es el de un solo proceso. Esto debido a que si se genera una regresión con estos datos, es el dato cuya existencia es la que introduce más error en esta regresión.

### 2. Crear un archivo de esos datos en el formato CSV y leerlo en un Dataframe de Pandas.

```python
df = pd.read_csv("datos_visinfo.csv")
df
```

### 3. Completar las columnas de Speeup, Efficiency y Cost usando las fórmulas mostradas en la diapositiva 6, tanto para el caso absoluto como para el caso relativo con p' = 16. Agregar estos datos al dataframe antes creado y realizar los gráficos correspondientes usando Matplotlib.

```python
speedup = []
efficiency = []
cost = []

speedup_r = []
efficiency_r = []
max_r_accel = []
p_prima = 16

t_1 = df['Time [s]'].iloc[0]
t_pprima = df['Time [s]'].iloc[1]
for index, row in df.iterrows(): 
    speedup.append(t_1/row['Time [s]'])
    efficiency.append(speedup[index]/row['Proc'])
    cost.append(row['Proc']*row['Time [s]'])
    speedup_r.append(t_pprima/row['Time [s]'])
    efficiency_r.append(p_prima*speedup_r[index]/row['Proc'])
    max_r_accel.append(row['Proc']/p_prima)

df['Speedup'] = speedup
df['Efficiency'] = efficiency
df['Cost'] = cost
df['Relative Speedup'] = speedup_r
df['Relative Efficiency'] = efficiency_r
df['Max. Rel. Acceleration'] = max_r_accel
```

```python
df
```

```python
x = df['Proc'].to_numpy()
y = df['Time [s]'].to_numpy()

fig, ax = plt.subplots()
ax.scatter(x, y)
m, b = np.polyfit(x, y, 1)
ax.plot(x,m*x+b, color='red',alpha=0.7)
ax.set(xlabel='Proc', ylabel='Time[s]',
       title='Time [s]')
ax.grid()
plt.show()


y = df['Speedup'].to_numpy()

fig, ax = plt.subplots()
ax.scatter(x, y)
m, b = np.polyfit(x, y, 1)
ax.plot(x,m*x+b, color='red',alpha=0.7)
ax.set(xlabel='Proc', ylabel='Speedup',
       title='Speedup')
ax.grid()
plt.show()

y = df['Efficiency'].to_numpy()

fig, ax = plt.subplots()
ax.scatter(x, y)
m, b = np.polyfit(x, y, 1)
ax.plot(x,m*x+b, color='red',alpha=0.7)
ax.set(xlabel='Proc', ylabel='Efficiency',
       title='Efficiency')
ax.grid()
plt.show()

y = df['Cost'].to_numpy()

fig, ax = plt.subplots()
ax.scatter(x, y)
m, b = np.polyfit(x, y, 1)
ax.plot(x,m*x+b, color='red',alpha=0.7)
ax.set(xlabel='Proc', ylabel='Cost',
       title='Cost')
ax.grid()
plt.show()

y = df['Relative Speedup'].to_numpy()

fig, ax = plt.subplots()
ax.scatter(x, y)
m, b = np.polyfit(x, y, 1)
ax.plot(x,m*x+b, color='red',alpha=0.7)
ax.set(xlabel='Proc', ylabel='Relative Speedup',
       title='Relative Speedup')
ax.grid()
plt.show()

y = df['Relative Efficiency'].to_numpy()

fig, ax = plt.subplots()
ax.scatter(x, y)
m, b = np.polyfit(x, y, 1)
ax.plot(x,m*x+b, color='red',alpha=0.7)
ax.set(xlabel='Proc', ylabel='Relative Efficiency',
       title='Relative Efficiency')
ax.grid()
plt.show()

y = df['Max. Rel. Acceleration'].to_numpy()

fig, ax = plt.subplots()
ax.scatter(x, y)
m, b = np.polyfit(x, y, 1)
ax.plot(x,m*x+b, color='red',alpha=0.7)
ax.set(xlabel='Proc', ylabel='Max. Rel. Acceleration',
       title='Max. Rel. Acceleration')
ax.grid()
plt.show()
```
