Implementación Figura/BoxPlot en formato clase

```python

class BoxPlot:
    """Clase BoxPlot
    
    Genera una figura BoxPlot con los detalles estéticos necesarios.


    """
    import matplotlib.pyplot as plt
    from matplotlib import rc
    import matplotlib.patheffects as path_effects
    def __init__(self,datos,colores_fill,colores_mediana,nombres,scale,titulo,label_x,label_y,limit_y,ax):
        """Constructor de BoxPlot
        
            __init__(self,datos,colores_fill,colores_mediana,nombres,scale,titulo,label_x,label_y,limit_y,ax)
            
            Se le pasan los datos/detalles estéticos necesarios para el boxplot.
            
            No devuelve nada.
        """
        self.datos=datos
        self.colores_fill = colores_fill
        self.colores_mediana = colores_mediana
        self.nombres = nombres
        self.scale = scale
        self.ax = ax
        self.titulo = titulo
        self.label_x = label_x
        self.label_y = label_y
        self.limit_y = limit_y
    def plot(self):
        """Constructor de BoxPlot
        
            plot(self)
            
            Utiliza los datos contenidos en la clase BoxPlot para graficar utilizando MatPlotLib.
            Esto permite actualizar la clase y solo volver a llamar a la función plot para así visualizar
            los cambios hechos.
            
            No devuelve nada.
        """
        ax = self.ax
        scale = self.scale
        nombres = self.nombres
        colores_mediana = self.colores_mediana
        colores_fill = self.colores_fill
        datos = self.datos
        label_y = self.label_y
        label_x = self.label_x
        limit_y = self.limit_y
        titulo = self.titulo
        whisker_shadow = False
        box_shadow = True
        text_shadow = True
        ax.get_xaxis().tick_bottom()
        ax.get_yaxis().tick_left()
        hay_notch = True
        bp = ax.boxplot(datos, patch_artist=True,notch=hay_notch)
        outline_color = 'xkcd:dark grey'
        fill_color = 'xkcd:white'
        median_color = 'xkcd:blueberry'
        bg_color = 'xkcd:silver'
        grid_color = 'xkcd:white'
        plot_shadow_offset = (scale//2,-scale//2)
        text_shadow_offset = (scale//4,-scale//4)
        plot_shadow_alpha = 0.2
        text_shadow_alpha = 0.2
        ax.grid(True, which='major', axis = 'y',color=bg_color, linestyle='-', linewidth=2,alpha=0.15)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['left'].set_visible(False)
        for idx in range(len(bp['boxes'])):
            box = bp['boxes'][idx]
            median = bp['medians'][idx]
            color_grupo = colores_fill[idx]
            box.set( color=outline_color, linewidth=2)
            box.set( facecolor = color_grupo,alpha=0.7)
            if box_shadow:
                box.set(path_effects=[path_effects.withSimplePatchShadow(offset=plot_shadow_offset,alpha=plot_shadow_alpha)])
            median.set(color=colores_mediana[idx], linewidth=2,alpha=0.7)
        for whisker in bp['whiskers']:
            whisker.set(color=outline_color, linewidth=2,alpha=0.7,linestyle='--')
            if whisker_shadow:
                whisker.set(path_effects=[path_effects.SimpleLineShadow(offset=plot_shadow_offset,alpha=plot_shadow_alpha),
                               path_effects.Normal()])
        for cap in bp['caps']:
            cap.set(color=outline_color, linewidth=2,alpha=0.7)
            if whisker_shadow:
                cap.set(path_effects=[path_effects.SimpleLineShadow(offset=plot_shadow_offset,alpha=plot_shadow_alpha),
                               path_effects.Normal()])
        for flier in bp['fliers']:
            flier.set(marker='o', color=outline_color, alpha=0.15,markerfacecolor=outline_color,
                          linestyle='none', markeredgecolor=outline_color)
        ax.set_xticklabels(nombres)
        ax.tick_params(axis=u'both', which=u'both',length=0)
        ax.xaxis.set_tick_params(labelsize=scale*2.5)
        ax.yaxis.set_tick_params(labelsize=scale*1.5)
        if limit_y:
            ax.set_ylim(limit_y)
        ax.set_title(titulo,fontsize= scale*5,path_effects=([path_effects.SimplePatchShadow(offset=text_shadow_offset,alpha=text_shadow_alpha),path_effects.Normal()] if text_shadow else []), pad=3*scale)
        ax.set_ylabel(label_y,fontsize = scale*3,path_effects=([path_effects.SimplePatchShadow(offset=text_shadow_offset,alpha=text_shadow_alpha),path_effects.Normal()] if text_shadow else []),labelpad=3*scale)
        ax.set_xlabel(label_x,fontsize = scale*3,path_effects=([path_effects.SimplePatchShadow(offset=text_shadow_offset,alpha=text_shadow_alpha),path_effects.Normal()] if text_shadow else []),labelpad=3*scale)
```

```python
class Figura:
    """Clase Figura
    
    Contiene varios BoxPlots, se puede accesar a cada uno y cambiar sus atributos.
    
    
    """
    import matplotlib.pyplot as plt
    from matplotlib import rc
    import matplotlib.patheffects as path_effects
    def plot(self,datos,colores_fill, colores_mediana, nombres,titulo,label_x,label_y,pos_x = 0, pos_y = 0,limit_y=None):
        """ plot(datos,colores_fill, colores_mediana, nombres,titulo,label_x,label_y,pos_x = 0, pos_y = 0,limit_y=None)
            
            Se le pasan los datos necesarios para dibujar un boxplot, incluyendo su posición en la Figura (x,y).
            Asimismo se le pasa el set de datos y otros detalles estéticos.
            
            No devuelve nada.
            
            TODO: Hay que generar más opciones en el constructor para sombras, etc.
        """
        bp = BoxPlot(datos,colores_fill,colores_mediana,nombres,self.escala,titulo,label_x,label_y,limit_y,self.axs[pos_y,pos_x])
        bp.plot()
    def __init__(self,figuras_x=1,figuras_y=1,escala = 6,titulo=''):
        """Constructor de figura
            
            Se le pasan las figuras en x y en y que se desea contener, así como una escala (para tamaño)
            y el título de la figura.
            
            No devuelve nada.
            
        """
        self.plt.rcParams["figure.figsize"] = (figuras_x*scale*1.1,figuras_y*scale)
        self.figuras_x = figuras_x
        self.figuras_y = figuras_y
        self.escala = escala
        self.titulo = titulo
        self.fig, self.axs = self.plt.subplots(figuras_y, figuras_x, squeeze=False)#,title=titulo)
        self.fig.suptitle(titulo, fontsize=escala*8)
```
