```python
from datetime import timedelta

def fillHoles(row):
    """
    Lo uso para arreglar huecos en los números de pruebas. Como son cumulativas, si hay días que hay cero
    pruebas claramente es un error, por lo que se encarga de hacer que el conjunto de datos esté bien formado
    """
    if np.isnan(row['Tests']) or row['Tests'] == 0.0:
        date = row['Date']
        country = row['Country']
        query = df_pruebas_caso.loc[(df_pruebas_caso['Country'] == country) & (df_pruebas_caso['Date'] == date - timedelta(days=1))]
        if query.empty == True:
            return 0
        else:
            q_value = query.reset_index().drop('index',axis=1).loc[0]['Tests']
            return q_value
    else:
        return row['Tests']


for idx, row in df_pruebas_caso.iterrows():
    df_pruebas_caso.at[idx, 'Tests'] = fillHoles(row)
```

```python
#Función que genera una lista de fechas en formato de texto en base a los días desde primer caso
def generaFechas(lista):
    locale.setlocale(locale.LC_TIME, 'es_ES.UTF-8')
    return [(start_date+timedelta(days=x)).strftime("%d %b") for x in lista]

#Función base que genera un gráfico de los datos relevantes a la cruva pruebas por caso
#Inspiración tomada de las gráficas de FiveThirtyEight
def plotCountry(country, xlim=False, ylim = False, plotTests = True, plotCases = True, fileName = None):
    fig = plt.gcf()
    fig.set_size_inches(18.5, 10.5)
    style.use('fivethirtyeight')
    colors = ["#d95f02", "#1b9e77", "#7570b3"]
    
    custom_cycler = (cycler(color=colors))
    
    df = df_p_c[df_p_c.Country.isin([country])].groupby('Date').sum()
    value_vars = ['TestsPerCase']
    if plotTests:
        value_vars.append('Tests')
    if plotCases:
        value_vars.append('Cases')
    df = pd.melt(df.reset_index(),
               id_vars=['Date'], value_vars = value_vars, value_name='Measure', var_name='Type')
    minday = df['Date'].min()
    maxday = df['Date'].max()
    minval = np.log10(df['Measure'].where(df['Measure'] > 0)).min()
    maxval = np.log10(df['Measure'].max())
    lista_ceros_pruebas = df.index[(df['Measure'] == 0) & (df['Type'] == 'Tests')].tolist()
    lista_ceros_tpc = df.index[(df['Measure'] == 0) & (df['Type'] == 'TestsPerCase')].tolist()
    if len(lista_ceros_pruebas) > 0:
        df.loc[lista_ceros_pruebas[:-1],'Measure']=None
        df.loc[lista_ceros_pruebas[-1],'Measure']=10**minval
        df.loc[lista_ceros_tpc[:-1],'Measure']=None
        df.loc[lista_ceros_tpc[-1],'Measure']=10**minval
        df = df.dropna()
    #mpl.rcParams['axes.prop_cycle'] = custom_cycler
    plt.rc('axes', prop_cycle=custom_cycler)
    
    g = sns.lineplot(x="Date", y="Measure", hue="Type", data=df,  legend = False)
    g.xaxis.set_major_locator(plt.MaxNLocator(8))
    
    g.lines[0].set_linewidth(6)
    g.lines[1].set_linestyle("--")
    g.lines[2].set_linestyle("--")
    
                     
    g.set(yscale="log")
    colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
    g.tick_params(axis = 'both', which = 'major', labelsize = 18)
    g.axhline(y = 10**minval, color = 'black', linewidth = 1.7, alpha = 0.7)
    g.set_xlim(left = minday-1, right = maxday)
    ticks_x = g.get_xticks()
    delta_ticks = ticks_x[1]-ticks_x[0]
    labels_formateados = generaFechas(ticks_x)
    g.set_xticklabels(labels_formateados)
    g.xaxis.label.set_visible(False)
    g.yaxis.label.set_visible(False)


    xpos_titulo = ticks_x[0]+delta_ticks/3
    
    return g, colors, minval, maxval, xpos_titulo
```

```python
#Función no utilizada todavía, permite ver como cambian los datos día a día
def generateGradient(values, percent = True):
    zer = np.array([0])
    if percent == True:
        return (np.append(values.to_numpy(),zer)- np.append(zer,values.to_numpy()))[:-1]/values.to_numpy()*100
    return (np.append(values.to_numpy(),zer)- np.append(zer,values.to_numpy()))[:-1]
```

```python
lista_dias = list(range(last_day+1))
lista_fechas = generaFechas(lista_dias)



def generaDatasetGradientes(paises, tam_ventana = 7):
    lista = []
    for p in paises:
        df_pais = df_p_c[df_p_c['Country'] == p]
        gradiente = generateGradient(df_pais['TestsPerCase'])
        df_pais['Gradient'] = np.convolve(gradiente, np.ones((tam_ventana,))/tam_ventana, mode='same')
        lista.append(df_pais)
    df_previo = pd.concat(lista)
    df_previo['Date'] = df_previo['Date'].astype(str)
    return  df_previo.pivot_table(index='Date', columns='Country', values='Gradient').fillna(0.0)
    
def getFechasConEspacios(lista):
    lista_salida = []
    for i, d in enumerate(lista):
        if i % 7 == 0:
            lista_salida.append(d)
        else:
            lista_salida.append("")
    return lista_salida
    
data = generaDatasetGradientes(paises_interes)
data
```
